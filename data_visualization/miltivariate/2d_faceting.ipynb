```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

df = pd.read_csv('fuel_econ.csv')
df['trans_type'] = df['trans'].apply(lambda x: x.split()[0])
df.head()
```

```python
g = sns.FacetGrid(data=df, col = 'VClass')
g.map(plt.scatter, 'displ', 'comb');
```

```python
g = sns.FacetGrid(data=df, col = 'VClass', row='trans_type',
                 margin_titles=True)
g.map(plt.scatter, 'displ', 'comb');
```
