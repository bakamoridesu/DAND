```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

df = pd.read_csv('pokemon.csv')
df.head()
```

```python
pkmn_stats = ['hp', 'attack', 'defense', 'speed', 'special-attack', 'special-defense']
g = sns.PairGrid(data=df, vars=pkmn_stats)
g = g.map_offdiag(plt.scatter)
g.map_diag(plt.hist);
```

```python
df[pkmn_stats].corr()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df[pkmn_stats].corr(), cmap='Reds', annot=True,
           fmt='.2f', vmin=0);
```
