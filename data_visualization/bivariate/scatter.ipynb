```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('fuel_econ.csv')
df.head()
```

```python
plt.scatter(data=df, x='displ', y='comb')
plt.xlabel('Displacement (l)')
plt.ylabel('Combined Fuel eff(mpq)');
```

```python
sns.regplot(data=df, x='displ', y='comb')
plt.xlabel('Displacement (l)')
plt.ylabel('Combined Fuel eff(mpq)');
```

```python
def log_trans(x, inverse = False):
    if not inverse:
        return x**2
    else:
        return x**0.5
```

```python
sns.regplot(df['displ'], df['comb'].apply(log_trans))
```

### Jitter, transparency

```python
sns.regplot(data=df, x='year', y='comb');
```

```python
sns.regplot(data=df, x='year', y='comb', x_jitter=0.3);
```

```python
sns.regplot(data=df, x='year', y='comb', x_jitter=0.3, scatter_kws={'alpha' : 1/5});
```
