```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('fuel_econ.csv')
df.head()
```

```python
g = sns.JointGrid(data=df, x='displ', y='comb')
g.plot_joint(plt.scatter)
g.plot_marginals(sns.distplot);
```

```python
plt.figure(figsize=(12,8))
sns.violinplot(data=df, x='VClass', y='comb', color='gray', inner='stick')
plt.xticks(rotation=20);
```

```python
plt.figure(figsize=(12,8))
sns.violinplot(data=df, x='VClass', y='comb', color='gray', inner='point')
plt.xticks(rotation=20);
```
